# Extract feature

```python
train_x = np.empty([12 * 471, 18 * 9], dtype = float)
train_y = np.empty([12 * 471, 1], dtype = float)

for month in range(12):    
    for day in range(20):
        for hour in range(24):
            if day*24+hour<471:
                train_x[month*471+day*24+hour,:] =  month_data[month][:,day*24+hour:day*24+hour+9].reshape(1,-1)
                train_y[month*471+day*24+hour,0] =  month_data[month][9,day*24+hour+9]            
print(train_x)
print(train_y)
```

# Normalization 

```python
train_x_mean=np.mean(train_x,axis=0)
train_x_std=np.std(train_x,axis=0)

for i in range(train_x.shape[0]):
    for j in range(train_x.shape[1]):
        train_x[i,j]= (train_x[i,j]-train_x_mean[j])/train_x_std[j]
        
train_x_mean
train_x_std
```

# Split data into training set and validation set

```python
import math
x_train_set = train_x[: math.floor(len(train_x) * 0.8), :]
y_train_set = train_y[: math.floor(len(train_y) * 0.8), :]
x_validation = train_x[math.floor(len(train_x) * 0.8): , :]
y_validation = train_y[math.floor(len(train_y) * 0.8): , :]
train_x.shape
x_train_set.shape
```

```python
dim = 18*9+1
w=np.zeros([dim,1])
x=np.concatenate((np.ones([12*471,1]),train_x), axis = 1).astype(float)#
#   np.concatenate((np.ones([12*471,1]),      x), axis = 1).astype(float)
import os 

learning_rate= 100
iteration_time= 1000
adagrad = np.zeros([dim,1])
for i in range(iteration_time):
    gradient = 2 * np.dot(x.transpose(),np.dot(x,w)-train_y)
    adagrad += gradient**2    
    w = w - learning_rate * gradient / np.sqrt(adagrad + 0.0000000001)
    
#     print(gradient)
#     print(adagrad)
#     print(i)
#     print("W")
#     print(w)
#     print("\n")
    

np.save('weight.npy',w)

```
